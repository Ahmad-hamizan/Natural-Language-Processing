```python
name = "Ahmad Hamizan"
age = 18
height = 165
weight = 45
is_student = True
```

list -> Mutable (Bisa diubah) []

tuple -> Immutable (Tidak bisa diubah) ()

dictionary -> Mutable {}

Tipe Data Collection

```python
# List
friends = ["Alice", "Bob", "Charlie"]
variable_with_list = [name, age, height, weight, is_student]
```

```python
variable_with_list.append("NLP")
```

```python
# Tuple
friends = ("Alice", "Bob", "Charlie")
```

```python
# Dictionary
my_self = {
    "name": name,
    "age": age,
    "height": height,
    "weight": weight,
    "is_student": is_student,
}
my_self.keys()
```

```python
my_self.values()
```

Looping/Perulangan

```python
for i in friends:
    print(i)
```

```python
# Perulangan dengan While

index = 0
while index < len(friends):
    print(friends[index])
    index += 1
```

function (fungsi)

```python
def introduce(name, age, height):
    print(name)
    print(age)
    print(height)
```

```python
introduce("Hamizan", 18, 165)
```

```python
number = 10
```

```python
if number % 2 == 0:
    print("Hebat Kamu Bud")
else:
    print("Apa-apain kek")
```
